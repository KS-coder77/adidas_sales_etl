```python
def check_data(df):
    summary = [
        [col, df[col].dtype, df[col].count(), df[col].nunique(), df[col].isnull().sum(), df.duplicated().sum()]
        for col in df.columns] 
    
    df_check = pd.DataFrame(summary, columns = ['column', 'dtype', 'instances', 'unique', 'missing_vals', 'duplicates'])
    
    return df_check 
```

```python
#function to change date columns to datetime 

def convert_to_datetime(df, col_name):
    try: 
        df[col_name] = pd.to_datetime(df[col_name], format = "%Y%m%d")
        return df
    except Exception as e:
        print(f"Error: {e}")
        return None 
```

```python
convert_to_datetime(data, 'Invoice Date')
```

```python
# add day of week, month and year cols 

def add_day_month_year_cols(df, col_name):
    df['Day of Week'] = df[col_name].dt.dayofweek
    df['Day Name'] = df[col_name].dt.day_name()
    df['Month'] = pd.to_datetime(df[col_name], format = '%m').dt.month
    df['Year'] = pd.to_datetime(df[col_name], format = '%Y').dt.year
    
```

```python
add_day_month_year_cols(data, 'Invoice Date')
```

```python
import sqlite3
import os, sys

#create db filepath 

DATABASE_FILEPATH = r"C:\temp\test102.db"

if os.path.exists(DATABASE_FILEPATH):
    os.remove(DATABASE_FILEPATH)

conn = sqlite3.connect(DATABASE_FILEPATH)
cur = conn.cursor()

cur.executescript("""

-- # 1. Date 
-- # cols: invoice date, year, month, day of week and day name, and date (yy-mm)

CREATE TABLE IF NOT EXISTS sales_date(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    inv_date TEXT, 
    year INT, 
    month INT, 
    day_of_week INT, 
    day_name TEXT
    );

-- # 2. Markets 
-- # cols: region, state, city

CREATE TABLE IF NOT EXISTS sales_markets(
    id INTEGER PRIMARY KEY AUTOINCREMENT, 
    region TEXT, 
    state TEXT, 
    city TEXT
    );

-- # 3. Products 
-- # cols: product_code (create new), product_type 

CREATE TABLE IF NOT EXISTS sales_products(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_type TEXT 
    );

-- # 4. sales_retailer 
-- # cols: retailer, retailer ID

CREATE TABLE IF NOT EXISTS sales_retailers(
    id INTEGER PRIMARY KEY AUTOINCREMENT, 
    retailer_name TEXT, 
    retailer_id INT
    );

-- # 5. transactions 
-- # cols: retailer ID, invoice date, product_type, price_per_unit, units_sold, total_sales, op_profit, 
--op_margin, sales_method

CREATE TABLE IF NOT EXISTS sales_transactions(
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    inv_date TEXT,
    retailer_id INT, 
    product_type TEXT,
    price_per_unit INT, 
    units_sold INT,
    total_sales INT,
    op_profit INT, 
    op_margin INT, 
    sales_method TEXT    
    )
    
    """)

# insert data into SQL table 
for index, row in data.iterrows():
    cur.execute('''
        INSERT INTO sales_date (inv_date, year, month, day_of_week, day_name)
        VALUES (?, ?, ?, ?, ? )''', 
        (row['Invoice Date'], row['Year'], row['Month'], row['Day of Week'], row['Day Name']));
        
    cur.execute('''
        INSERT INTO sales_markets (region, state, city) 
        VALUES (?, ?, ?)''',
        (row['Region'], row['State'], row['City']));
    
    cur.execute('''
        INSERT INTO sales_products (product_type) 
        VALUES (?)''',
        (row['Product'], ));
    
    cur.execute('''
        INSERT INTO sales_retailers (retailer_name, retailer_id)
        VALUES (?, ?)''',
               (row['Retailer'], row['Retailer ID']))
    
    cur.execute('''
        INSERT INTO sales_transactions (inv_date, retailer_id, product_type, price_per_unit, units_sold, total_sales, op_profit, op_margin, sales_method) 
        VALUES (?, ?, ?, ?, ?, ?, ?, ?, ? )''',
        (row['Invoice Date'], row['Retailer ID'], row['Product'], row['Price per Unit'], row['Units Sold'], row['Total Sales'], row['Operating Profit'], row['Operating Margin'], row['Sales Method']))
    
      
# commit changes and close the connection 
conn.commit()
#conn.close()
```

```python
date_df=pd.read_sql('SELECT * from invoice_date', conn)
date_df.to_csv('inv_date.csv', index=False)
```

```python
markets_df=pd.read_sql('SELECT * from markets', conn)
markets_df.to_csv('markets.csv', index=False)

products_df=pd.read_sql('SELECT * from products', conn)
products_df.to_csv('prods.csv', index=False)

trans_df=pd.read_sql('SELECT * from transactions', conn)
trans_df.to_csv('trans.csv', index=False)
```
